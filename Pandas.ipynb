#Q9. How do you concatenate two Pandas DataFrames?

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"]
    },index=[0,1,2,3])
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B2", "B3"]
    },index=[4, 5, 6, 7]
)
df3 = pd.concat([df1,df2])
df3
```

#Q10. How do you merge two Pandas DataFrames on a specific column?

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3","K4"],
        "A": ["A0", "A1", "A2", "A3","A4"],
        "B": ["B0", "B1", "B2", "B3","B4"],
    }
)
right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
result = pd.merge(left, right, how="inner",on="key")
result
```
